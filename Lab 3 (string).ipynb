```python
#Task 10 (Ei niyom dekha lagbe na, porer duita OR1, OR2 dekhbo. )

word = str(input())

firstPart = ""
lastPart = ""

flagFirstPart = True
for i in word:
  if i == "," or i == " ":
    flagFirstPart = False
    continue
  
  if flagFirstPart == True:
    firstPart = firstPart + i
  else:
    lastPart = lastPart + i

print(firstPart) # ei print duita sample output wise korar dorkar nei
print(lastPart)  # code er first part e ki korlam ta bujhar jonno print korsi.

bigLength = 0
if len(firstPart) >= len(lastPart):
  bigLength = len(firstPart)
else:
  bigLength = len(lastPart)


finalWord = ""

for i in range(0, bigLength, 1):
  fc = firstPart[i:i+1:]
  lc = lastPart[i:i+1:]

  finalWord = finalWord + fc + lc

print(finalWord)

```

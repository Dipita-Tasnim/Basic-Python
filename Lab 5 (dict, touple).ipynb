```python
# Task 11 (correct)...........assign

# frequency of character: character gulo koybar kore ache sheta bujhay.

s = input()

d = {}
s1 = ''

for i in s:
  if i >= 'A' and i <= 'Z' :
    ch = chr(ord(i)+32)  # +32 korle small letter pawa jay.
    s1 = s1 + ch         # orthat user boro haat, choto haat mixed kore dileo
                         # output e hishab shob choto haat e hobe,shob choto show korbe.

  else:
    s1 = s1 + i  # ei condition dite bhule jabo na. karon user number or special charc
                 # mix kore dile shegulo o jate new string e included thake.
# Q. wise P , p duita ke same consider korar permission ache. tai uporer part na
# korleo problem nai. tarpor o jene rakha valo. small,capital jekono ekta te nilei hoy.

for i in s1:
  if i == ' ' :
    continue  # pass dile hobe na. pass deya mane actually kichui na kora. tai
              # pass dile jehetu if e dhukse, baki 2 ta else e jabe na..
              # and space ta include hobe.
              # aar continue dile kono dike na takay loop e jabe and s er next element
              # ta i nibe.. tarpor oita diye check kora start korbe. so, space include
              # holo na.

  elif i in d:
    d[i] = d[i] + 1

  else:
    d[i] = 1

print(d)


# ekhane ekdom first e jokhon loop ghurlo tokhon i e p ashlo jeta d te nei karon
# first e d faka. tai p er val 1(else e gese). thenloop ghurte ghurte jokhon
# programming er p ashlo tokhon d te aage already p thakay p er val update
# hoye 2 hoyeche.
# ekarone output e ekdom first p er val 2. aar update er karonei same charc output
# e barbar repeat hoy nai.

```

```python
# TRACING

dict1 = {'a':59 , 'b':-82 , 'c':5 , 'd':-81 , 'e':53}
for i in dict1:
  j = 0
  k = 22
  while j < 5:
    if j % 2 == 0:
      k = dict1[i] + j - (8 + k % 6) / 3
      dict1[i] = dict1[i]+ int(k)
    else:
      k = dict1[i] + j - (6 - k % 8) * 3
      dict1[i] = dict1[i] - int(k)
    j += 1
  print(int(k))
  print(i + " -> " + str(dict1[i]))

```

```python
## task 8 (correct)


def show_palindrome(num):
  s = ''

  for i in range(1, num + 1):
    s = s + str(i) + ' '

  for i in range(num-1, 0, -1):
    s = s + str(i) + ' ' 

  return s


def show_palindrome_triangle(num):
  for i in range(1, num + 1):
    for j in range(1, num - i + 1):
      print('  ',end ='')    # ekhane duita space

    print(show_palindrome(i))

show_palindrome_triangle(5)           

```

```python
# Task 13
#Ekta function er bhitor onekgulo return dite hole if elif use kore kore return dite hobe.
#if elif deya mane shobgulo condition check hobe but dhukbe just ektay..aar oitar bhitor 
# ekta return thakbe jetar kaaj hobe..return korar por nicher baki elif gulor kaaj bondho
# holeo problem nai karon input condition wise amar kaaj shesh.
# if elif use na korle shurur addition tai shudhu print korbe jei operator ee input e dei.
# addition er return er karone nicher shob code okejo hoye jabe.

operator = input()
first_operand = float(input())
second_operand = float(input())

def function_name(operator, first_operand, second_operand):

  if operator == '+':
    addition = first_operand + second_operand
    return addition

  elif operator == '-':  
    subtraction = first_operand - second_operand
    return subtraction

  elif operator == '*':  
    multiplication = first_operand * second_operand
    return multiplication

  elif operator == '/':  
    division = first_operand / second_operand
    return division

print(function_name(operator, first_operand, second_operand))   
```

```python
# task 14

sentence = input()
position = int(input())

def function_name(sentence, position):

  s1 = ''
  s2 = ''
  new_s = ''

  for i in range(len(sentence)):
    if i % position != 0 or i == 0:
      s1 = s1 + sentence[i]

    elif i % position == 0 and i != 0:
      s2 = s2 + sentence[i]

  new_s = s1 + s2
  return new_s

print(function_name(sentence, position))   

```

```python
#Task 6 (Eta dekhbo na..porer ta dekhbo)

s=input()
s=s[1:-1]  ## dui pasher inverted comma baad diye majher element gulo rakha hoise.
           # inveretd comma o ekekta element jeta 0 and -1 no. index e ache.
s=s.split(', ')
lst1=[]

for i in s:
  lst1.append(int(i))

print('My list:', lst1)  
  
max=int(lst1[0])
pos=0

for i in range(len(lst1)):
  val=int(lst1[i])
  if val>max:
    max=val # max e first ei 0 thakbe then i er value dara update hote thakbe.
    pos=i   # i er index number dara position update hote thakbe.
    
print('Largest number in the list is',max, 'which was found at index', pos)
  
```

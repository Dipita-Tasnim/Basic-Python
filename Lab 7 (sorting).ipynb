```python
## Slection sort (purotar shathe check kore)

lst = [7, 100, 3, 10, -4]

for idx in range(0, len(lst)-1):
  min_val = lst[idx]
  min_idx = idx

  for j in range(idx+1, len(lst)):
    if lst[j] < min_val:
      min_val = lst[j]
      min_idx = j

  lst[idx], lst[min_idx] = lst[min_idx], lst[idx]

  # temp = lst[idx]
  # lst[idx] = lst[min_idx]
  # lst[min_idx] = temp

print(lst)  
```

```python
## Bubble sort (pashapashi takay)

lst = [7, 5, 3, 4, 1]

for i in range(len(lst)-1):
  for j in range(0, len(lst)-i-1):
    if lst[j] > lst[j+1]:
      temp = lst[j]
      lst[j] = lst[j+1]
      lst[j+1] = temp
print(lst)
```

```python
## linear (puratar shathe check kore)

lst = [2,4,9,12,15,23,38,56]

def linear_search(lst, x):
  if x not in lst:
    return 'Not Found'

  else:
    for i in range(len(lst)):
      if lst[i] == x:
        return i

print(linear_search(lst, 23))          
```

```python
# Task 4

lst = [10,30,20,70,11,15,22,16,58,100,12,56,70,80]
even_lst=[]
odd_lst=[]

for i in range(0,len(lst),2):
  even_lst += [lst[i]] 

for i in range(1, len(lst), 2): 
  odd_lst += [lst[i]] 


for idx in range(0, len(even_lst)-1):
  min_val = even_lst[idx]
  min_idx = idx

  for j in range(idx+1, len(even_lst)):
    if even_lst[j] < min_val:
      min_val = even_lst[j]
      min_idx = j

  even_lst[idx], even_lst[min_idx] = even_lst[min_idx], even_lst[idx]


for idx in range(0, len(odd_lst)-1):
  min_val = odd_lst[idx]
  min_idx = idx

  for j in range(idx+1, len(odd_lst)):
    if odd_lst[j] > min_val:
      min_val = odd_lst[j]
      min_idx = j

  odd_lst[idx], odd_lst[min_idx] = odd_lst[min_idx], odd_lst[idx]


final_lst = []

for k in range(len(even_lst)):
  final_lst = final_lst + [even_lst[k]] + [odd_lst[k]]

print(final_lst)
```
